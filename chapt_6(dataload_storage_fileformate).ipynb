```python
# 迭代处理ex6.csv 将值计数聚合到‘key’列中
chunker = pd.read_csv(curDir+'ex6.csv',chunksize=1000)
tot = Series([])
for piece in chunker:
    tot = tot.add(piece['key'].value_counts(),fill_value=0)
tot = tot.sort_values(ascending = False)
tot[:10]
```

```python
# python 内置csv模块
import csv
f = open(curDir+'ex7.csv')
reader = csv.reader(f)
for line in reader:
    print(line)
```

```python
# 为使数据格式合乎要求，需要做一些整理工作
lines = list(csv.reader(open(curDir+'ex7.csv')))
header,values = lines[0],lines[1:]
data_dict = {h: v for h, v in zip(header,zip(*values))}
data_dict
```

```python
# csv文件的形式很多，只需定义csv.Dialect的一个子类即可定义出新格式
# （如专门的分隔符、字符串引用约定、行结束符等）
csv.register_dialect('mydialect',delimiter=',',lineterminator ='\n',quotechar='"')# 注册类
reader1 = csv.reader(f,'mydialect')
# 不用Dialect子类
reader2 = csv.reader(f,delimiter='|')
reader1
```

```python
with open(curDir+'mydata.csv','w') as f:
    writer = csv.writer(f,'mydialect')
    writer.writerow(('one','two','three'))
    writer.writerow(('1','2','3'))
    writer.writerow(('4','5','6'))
```
